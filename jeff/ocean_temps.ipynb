```python
import pandas as pd
import datetime as dt
```

```python
# The path to our CSV file
file = "Resources/GlobalTemperatures.csv"

# Read our global land and ocean temperature data into pandas
df = pd.read_csv(file)

df.head()
```

```python
# Convert dt field to datetime and delete records before 1900
df['dt'] = pd.to_datetime(df['dt'])
gt_df = df[(df['dt'].dt.year >= 1900)]

gt_df.head()
```

```python
# Assign columns more manageable names
gt_df.rename(columns={'dt': 'Date', 'LandAverageTemperature': 'LandAvgTemp',\
    'LandAverageTemperatureUncertainty': 'LandAvgUncert', 'LandMaxTemperature': 'LandMaxTemp',\
    'LandMaxTemperatureUncertainty': 'LandMaxUncert', 'LandMinTemperature': 'LandMinTemp',\
    'LandMinTemperatureUncertainty': 'LandMinUncert', 'LandAndOceanAverageTemperature': 'LandOceanAvgTemp',\
    'LandAndOceanAverageTemperatureUncertainty': 'LandOceanUncert'}, inplace=True)

gt_df.head()
```

```python
# Drop any records without ocean temperature data
gt_df.dropna(subset=['LandOceanAvgTemp', 'LandOceanUncert'])

gt_df.head()
```

```python
# Set date as index
gt_df.set_index('Date', inplace=True)

gt_df.head()
```

```python
gt_df.to_csv('./Resources/ocean_temps.csv')
```
